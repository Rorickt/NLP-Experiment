```python
from drive.MyDrive.reprod_code.hatebert_tune_code import run_hatebert
import pandas as pd

# Seed values used in the present experiment [42,53, 64, 75, 86, 97, 12,13,24,54,62,37]
# they were generated randomly using np.random (between 0 and 100) in a seperate file
seed_list = [42, 53, 64, 75, 86, 97, 12, 13, 24, 54, 62, 37]
```

```python
# Fine-tune HateBERT on each seed value and store the results per epoch
for seed_val in seed_list:
    f_scores, runs, epochs_num = run_hatebert(seed_val, 5)
    run_fscores = pd.DataFrame()
    run_fscores['run'] = runs
    run_fscores['epoch'] = epochs_num
    run_fscores['f1-score'] = f_scores
    run_fscores.to_csv(f'/content/drive/MyDrive/reprod_code/results/run_fscores{seed_val}.csv')
    run_fscores.to_csv(f'run_fscores{seed_val}.csv')
```
